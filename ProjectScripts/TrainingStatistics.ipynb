```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../all_stats.txt')
```

```python
plt.plot( [(i+1) for i in range(len(data['Accuracy']))], data['Accuracy'], label = 'Accuracy', color='green')
plt.plot( [(i+1) for i in range(len(data['Loss']))], data['Loss'], label= 'Loss', color="red")
plt.xlabel("Iteration")
plt.ylabel("Accuracy/Loss")
plt.title("Loss and accuracy per iteration")
plt.legend()
plt.show()
```
